```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Admission_Predict.csv")
```

```python
df.sample(7)
df.shape
```

```python
df = df.drop(columns = ["Serial No."])
```

```python
df.head()
```

```python
x = df.drop(columns = ["Chance of Admit "])
y = df["Chance of Admit "]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(12,activation='relu',input_dim=7))
model.add(Dense(6,activation='relu'))
model.add(Dense(1,activation='relu'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,batch_size=100,epochs=100,verbose=1)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred)*100)
print(100 - mean_squared_error(y_test,y_pred)*100)
```
