```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train/255
# y_train = y_train/255
```

```python
X_train = X_train.reshape(-1,28,28,1)
X_test = X_test.reshape(-1,28,28,1)
```

```python
model = Sequential()
```

```python
model.add(Conv2D(64, kernel_size=(3,3), padding="valid", strides=(2,2), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid'))
model.add(Conv2D(32, kernel_size=(3,3), padding="valid", strides=(2,2), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid'))
```

```python
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,batch_size=1500,epochs=64,verbose=1,validation_data=(X_test, y_test))
```

```python
y_pred = model.predict(X_test)
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred.argmax(axis=1))
```
